```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 2: Create your real-life dataset ---
data = pd.DataFrame({
    "Date": ["2025-10-29", "2025-10-30", "2025-10-31", "2025-11-01", "2025-11-02", "2025-11-03", "2025-11-04"],
    "Sleep": [90, 92, 89, 91, 87, 88, 90],
    "Recovery": [86, 88, 83, 85, 84, 90, 92],
    "Strain": [8.7, 7.2, 5.8, 9.0, 7.5, 6.9, 8.0],
    "Nutrition": [8, 9, 8, 9, 7, 8, 9],
    "Focus": [8, 9, 7, 8, 6, 9, 9],
    "Emotional": [7, 8, 7, 9, 6, 8, 9],
    "Productivity": [82, 90, 75, 88, 70, 85, 92]
})

print("ðŸ“Š Dataset preview:")
display(data)

# --- Step 3: Correlation map ---
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Between Habits and Productivity")
plt.show()

# --- Step 4: Insights ---
print("\nâœ¨ Insights:")
print(f"Average productivity: {data['Productivity'].mean():.1f}")
print(f"Best recovery day: {data.loc[data['Recovery'].idxmax(), 'Date']}")
print(f"Highest focus day: {data.loc[data['Focus'].idxmax(), 'Date']}")

# --- Step 5: Visualization over time ---
plt.figure(figsize=(10,5))
plt.plot(data["Date"], data["Productivity"], marker='o', label='Productivity')
plt.plot(data["Date"], data["Sleep"], marker='x', label='Sleep')
plt.plot(data["Date"], data["Focus"], marker='s', label='Focus')
plt.legend()
plt.title("Neuroplastic Learning System â€“ Week 1")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
